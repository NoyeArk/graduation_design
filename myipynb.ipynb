```python
import pandas as pd
from collections import Counter
```

```python
file_path = 'datasets/processed/Kindle/ratings.data'
df = pd.read_csv(file_path, sep='\t', header=None, nrows=10000)
df.head()
```

```python
df.columns = ['user', 'item', 'rating', 'time']
df.head()
```

```python
df = df.sort_values(['user', 'time'])
df.head()
```

```python
u_i = df[['user','item']].values.tolist()
u_i
```

```python
u_i = [list(map(str, line)) for line in u_i]
u_i
```

```python
us, bs = zip(*u_i)
us, bs
```

```python
C_us = Counter(us)
C_us
```
