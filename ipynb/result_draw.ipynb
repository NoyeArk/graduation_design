```python
import numpy as np
import matplotlib.pyplot as plt

# 
hatches = ['/', '\\', '|', '-', '+', '']  # 

datasets = ['ML-1M', 'KuaiRec', 'Amazon']
methods = ['CombSUM', 'CombANZ', 'CombMNA', 'Stream', 'AEM', 'EnsRec']
metrics = ['ndcg@10', 'map@10', 'mrr@10']
metric_values = {
    'ndcg@10': {
        'ML-1M': [0.3914, 0.3682, 0.3995, 0.4186, 0.4116, 0.4518],
        'KuaiRec': [0.4072, 0.4174, 0.4353, 0.4256, 0.4313, 0.5136],
        'Amazon': [0.2306, 0.1297, 0.2298, 0.2409, 0.2429, 0.2847]
    },
    'map@10': {
        'ML-1M': [0.2662, 0.2299, 0.2726, 0.2867, 0.2851, 0.3219],
        'KuaiRec': [0.2931, 0.3155, 0.3052, 0.3032, 0.3100, 0.3809],
        'Amazon': [0.1577, 0.0808, 0.1571, 0.1680, 0.1740, 0.2148]
    },
    'mrr@10': {
        'ML-1M': [0.2836, 0.2888, 0.2922, 0.3077, 0.3100, 0.3594],
        'KuaiRec': [0.3284, 0.3553, 0.3435, 0.3311, 0.3357, 0.4283],
        'Amazon': [0.1600, 0.0839, 0.1594, 0.1715, 0.1771, 0.2364]
    }
}

fig, axes = plt.subplots(3, 3, figsize=(18, 12), sharey='row')
plt.rcParams.update({'font.size': 16})  # 16

for row, metric in enumerate(metrics):
    for col, dataset in enumerate(datasets):
        ax = axes[row, col]
        values = metric_values[metric][dataset]
        x = np.arange(len(methods))
        bars = ax.bar(x, values, facecolor='none', edgecolor='black')  # 
        for bar, hatch in zip(bars, hatches):  # 
            bar.set_hatch(hatch)
        ax.set_title(dataset)
        ax.set_ylabel(metric, fontsize=16)
        ax.set_xticks(x)
        ax.set_xticklabels(methods, rotation=45)
        ax.set_ylim(0, 0.6)

plt.tight_layout()
plt.show()
```
