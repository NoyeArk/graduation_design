```python
import json
from tqdm import tqdm
```

## 处理数据集

处理交互信息

```python
file = "D:\work_space\Office_Products.jsonl"
with open(file, 'r') as fp:
    for line in tqdm(fp):
        data = json.loads(line.strip())
        user_id = data['user_id']
        asin = data['asin'] 
        rating = data['rating']
        timestamp = data['timestamp']

        with open('D:/Code/graduation_design/data/Office_Products/interaction.dat', 'a') as f:
            f.write(f"{user_id}::{asin}::{rating}::{timestamp}\n")
```

处理物品信息

```python
file = "D:\work_space\meta_Office_Products.jsonl"
with open(file, 'r') as fp:
    for line in tqdm(fp):
        data = json.loads(line.strip())
        parent_asin = data['parent_asin']
        with open('D:/Code/graduation_design/data/Office_Products/item.dat', 'a') as f:
            f.write(f"{parent_asin}>>{json.dumps(data)}\n")
```

## 验证数据集

```python
import pandas as pd
df = pd.read_csv("D:/Code/graduation_design/data/Office_Products/interaction.dat", sep=",", header=None, engine='python')
df.columns = ['user', 'item', 'rating', 'timestamp']
df
```

```python
df['item'].value_counts()
```

```python
data = df[df['rating'] > 2]

item_counts = data['item'].value_counts()
items_to_keep = item_counts[item_counts >= 10].index
data = data[data['item'].isin(items_to_keep)]

user_counts = data['user'].value_counts()
users_to_keep = user_counts[user_counts >= 10].index
data = data[data['user'].isin(users_to_keep)]

print(f"用户数量: {len(data['user'].unique())}, 物品数量: {len(data['item'].unique())}")

data = data.sort_values(['user', 'timestamp'])
data = data.reset_index(drop=True)
data
```
