```python
import pandas as pd
```

```python
df = pd.read_csv("D:\Code\graduation_design\old_datasets\processed\Kindle\\ratings.data", sep="\t")
df.columns = ['user', 'item', 'rating', 'timestamp']
df
```

```python
data = df[df['rating'] > 3.5]

item_counts = data['item'].value_counts()
items_to_keep = item_counts[item_counts >= 50].index
data = data[data['item'].isin(items_to_keep)]

user_counts = data['user'].value_counts()
users_to_keep = user_counts[user_counts >= 25].index
data = data[data['user'].isin(users_to_keep)]

print(f"用户数量: {len(data['user'].unique())}, 物品数量: {len(data['item'].unique())}")

data = data.sort_values(['user', 'timestamp'])
data = data.reset_index(drop=True)
data
```

```python
data.to_csv("kindle_data.csv", index=False)
```
